```python
# Importa las bibliotecas necesarias
import matplotlib.pyplot as plt
import random

# Genera datos de producción para dos fábricas
fabrica1 = [random.uniform(50, 100) for _ in range(100)]
fabrica2 = [random.uniform(50, 100) for _ in range(100)]

# Calcula estadísticas descriptivas para fabrica1
media1 = sum(fabrica1) / len(fabrica1)
mediana1 = sorted(fabrica1)[len(fabrica1) // 2]
moda1 = max(set(fabrica1), key=fabrica1.count)

# Calcula estadísticas descriptivas para fabrica2
media2 = sum(fabrica2) / len(fabrica2)
mediana2 = sorted(fabrica2)[len(fabrica2) // 2]
moda2 = max(set(fabrica2), key=fabrica2.count)

# Crea subplots para los histogramas
plt.figure(figsize=(10, 6))

# Histograma para fabrica1
plt.subplot(1, 2, 1)
plt.hist(fabrica1, bins=20, edgecolor='black')
plt.xlabel('Metros producidos')
plt.ylabel('Número de telares')
plt.title(f'Fabrica 1: Media={media1:.2f}, Mediana={mediana1:.2f}, Moda={moda1}')

# Histograma para fabrica2
plt.subplot(1, 2, 2)
plt.hist(fabrica2, bins=20, edgecolor='black')
plt.xlabel('Metros producidos')
plt.ylabel('Número de telares')
plt.title(f'Fabrica 2: Media={media2:.2f}, Mediana={mediana2:.2f}, Moda={moda2}')

# Ajusta el espacio entre subplots
plt.tight_layout()

# Muestra la figura
plt.show()
```
