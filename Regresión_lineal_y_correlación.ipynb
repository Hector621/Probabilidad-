```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
from google.colab import drive

# Montar Google Drive
drive.mount('/content/drive')

# Cargar los datos desde Google Drive
df1 = pd.read_csv('/content/drive/MyDrive/grupo1.csv')
df2 = pd.read_csv('/content/drive/MyDrive/grupo2.csv')
df3 = pd.read_csv('/content/drive/MyDrive/grupo3.csv')

# Imprimir los nombres de las columnas
print("Columnas del Grupo 1:", df1.columns)
print("Columnas del Grupo 2:", df2.columns)
print("Columnas del Grupo 3:", df3.columns)

# Histograma de calificaciones para los tres grupos
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df1['Calificacion'], bins=10, kde=True)
plt.title('Histograma de Calificaciones - Grupo 1')
plt.xlabel('Calificación Final')
plt.ylabel('Frecuencia')

plt.subplot(1, 3, 2)
sns.histplot(df2['Calificacion'], bins=10, kde=True)
plt.title('Histograma de Calificaciones - Grupo 2')
plt.xlabel('Calificación Final')
plt.ylabel('Frecuencia')

plt.subplot(1, 3, 3)
sns.histplot(df3['Calificacion'], bins=10, kde=True)
plt.title('Histograma de Calificaciones - Grupo 3')
plt.xlabel('Calificación Final')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()

# Definir función para analizar cada grupo
def analizar_grupo(df, grupo):
    X = df[['HorasEstudio']]
    y = df['Calificacion']

    # Crear y entrenar el modelo de regresión lineal
    reg = LinearRegression()
    reg.fit(X, y)

    # Obtener la pendiente e intercepto
    pendiente = reg.coef_[0]
    intercepto = reg.intercept_

    # Calcular el coeficiente de correlación de Pearson
    correlacion, _ = pearsonr(df['HorasEstudio'], df['Calificacion'])

    # Mostrar resultados
    print(f"Grupo {grupo}")
    print(f"Pendiente: {pendiente}")
    print(f"Intercepto: {intercepto}")
    print(f"Coeficiente de correlación de Pearson: {correlacion}\n")

    # Graficar los datos junto con la línea de regresión
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='HorasEstudio', y='Calificacion', data=df)
    plt.plot(df['HorasEstudio'], reg.predict(X), color='red')
    plt.title(f'Regresión Lineal - Grupo {grupo}')
    plt.xlabel('Horas de Estudio')
    plt.ylabel('Calificación Final')
    plt.show()

# Analizar cada grupo
analizar_grupo(df1, 1)
analizar_grupo(df2, 2)
analizar_grupo(df3, 3)
```
