```python
# Probabilidad y estadística
# Facilitador: Jose Gabriel Rodriguez Rivas
# Alumno: Jesus Adrian Ramirez Pacheco

# Importar librerías necesarias
import random
import matplotlib.pyplot as plt
from statistics import mean, median, mode, stdev

# Generar datos para las dos fábricas
fabrica1 = [round(random.uniform(50, 100), 2) for _ in range(100)]
fabrica2 = [round(random.uniform(50, 100), 2) for _ in range(100)]

# Calcular medidas de tendencia central y dispersión para fábrica 1
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)
desviacion1 = stdev(fabrica1)

# Calcular medidas de tendencia central y dispersión para fábrica 2
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)
desviacion2 = stdev(fabrica2)

# Imprimir resultados
print("Valores de la fábrica 1")
print(f"La media es: {media1}")
print(f"La mediana es: {mediana1}")
print(f"La moda es: {moda1}")
print(f"La desviación estándar es: {desviacion1}\n")

print("Valores de la fábrica 2")
print(f"La media es: {media2}")
print(f"La mediana es: {mediana2}")
print(f"La moda es: {moda2}")
print(f"La desviación estándar es: {desviacion2}\n")

# Graficar histogramas
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

ax1.hist(fabrica1, bins=10, alpha=0.7, color='blue', edgecolor='black')
ax1.axvline(media1, color='red', linestyle='dashed', linewidth=2, label='Media')
ax1.axvline(mediana1, color='yellow', linestyle='dashed', linewidth=2, label='Mediana')
ax1.axvline(moda1, color='green', linestyle='dashed', linewidth=2, label='Moda')
ax1.set_title('Producción diaria de la fábrica 1')
ax1.set_xlabel('Producción (metros)')
ax1.set_ylabel('Frecuencia')
ax1.legend()

ax2.hist(fabrica2, bins=10, alpha=0.7, color='blue', edgecolor='black')
ax2.axvline(media2, color='red', linestyle='dashed', linewidth=2, label='Media')
ax2.axvline(mediana2, color='yellow', linestyle='dashed', linewidth=2, label='Mediana')
ax2.axvline(moda2, color='green', linestyle='dashed', linewidth=2, label='Moda')
ax2.set_title('Producción diaria de la fábrica 2')
ax2.set_xlabel('Producción (metros)')
ax2.set_ylabel('Frecuencia')
ax2.legend()

plt.tight_layout()
plt.show()
```
