```python
# Libreria para generar numeros aleatorios
import random
import matplotlib.pyplot as plt
from statistics import mean, median, mode

# Generar los datos de producción para las dos fábricas
fabrica1 = [round(random.uniform(50, 100), 2) for _ in range(100)]
fabrica2 = [round(random.uniform(50, 100), 2) for _ in range(100)]

# Calcular medidas de tendencia central para fabrica1
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

# Calcular medidas de tendencia central para fabrica2
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)

# Imprimir los resultados
print("Valores de la fábrica 1:")
print("Media =", media1)
print("Mediana =", mediana1)
print("Moda =", moda1)

print("\nValores de la fábrica 2:")
print("Media =", media2)
print("Mediana =", mediana2)
print("Moda =", moda2)

# Graficar histogramas
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Histograma para fábrica 1
ax1.hist(fabrica1, bins=10, alpha=0.7, color='b', edgecolor='black')
ax1.axvline(media1, color='r', linestyle='dashed', linewidth=2, label='Media')
ax1.axvline(mediana1, color='g', linestyle='dashed', linewidth=2, label='Mediana')
ax1.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label='Moda')
ax1.set_title('Producción diaria de la fábrica 1')
ax1.set_xlabel('Producción (metros)')
ax1.set_ylabel('Frecuencia')
ax1.legend()

# Histograma para fábrica 2
ax2.hist(fabrica2, bins=10, alpha=0.7, color='b', edgecolor='black')
ax2.axvline(media2, color='r', linestyle='dashed', linewidth=2, label='Media')
ax2.axvline(mediana2, color='g', linestyle='dashed', linewidth=2, label='Mediana')
ax2.axvline(moda2, color='y', linestyle='dashed', linewidth=2, label='Moda')
ax2.set_title('Producción diaria de la fábrica 2')
ax2.set_xlabel('Producción (metros)')
ax2.set_ylabel('Frecuencia')
ax2.legend()

plt.tight_layout()
plt.show()
```
