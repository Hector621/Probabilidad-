```python
# Import libraries
import random
from statistics import *

# Data generation
fabrica1 = []  # List to store production values for factory 1
fabrica2 = []  # List to store production values for factory 2

# Generate data for 100 days (assuming 2 factories)
for _ in range(100):
    fabrica1.append(round(random.uniform(50, 100), 2))
    fabrica2.append(round(random.uniform(50, 100), 2))

# Central tendency measures
print("**Fábrica 1:**")
print(f"Media: {mean(fabrica1)}")
print(f"Mediana: {median(fabrica1)}")
print(f"Moda: {mode(fabrica1)}")

print("\n**Fábrica 2:**")
print(f"Media: {mean(fabrica2)}")
print(f"Mediana: {median(fabrica2)}")
print(f"Moda: {mode(fabrica2)}")

# Install matplotlib if not already installed (Run this cell once)
# !pip install matplotlib

# Data visualization (requires matplotlib)
import matplotlib.pyplot as plt

# Create histograms
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.hist(fabrica1, bins=10, edgecolor='black', alpha=0.7)
plt.title('Fábrica 1')
plt.xlabel('Producción diaria')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
plt.hist(fabrica2, bins=10, edgecolor='black', alpha=0.7)
plt.title('Fábrica 2')
plt.xlabel('Producción diaria')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()

# Descriptive statistics (not executed)
# min_fab1 = min(fabrica1)
# max_fab1 = max(fabrica1)
# rango_fab1 = max_fab1 - min_fab1

# min_fab2 = min(fabrica2)
# max_fab2 = max(fabrica2)
# rango_fab2 = max_fab2 - min_fab2

# Standard deviation (not executed)
# std_dev_fab1 = stdev(fabrica1)
# std_dev_fab2 = stdev(fabrica2)

# Histograms with standard deviation (not executed)
# # ... (code to create histograms with standard deviation lines)
```
