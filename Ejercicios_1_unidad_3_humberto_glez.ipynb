```python
import matplotlib.pyplot as plt

# Datos de ejemplo
momentos_soleados = [0, 1, 2, 3]
frecuencia_relativa = [0.1, 0.1, 0.4, 0.4]  # Frecuencia relativa para cada cantidad de momentos soleados

# Crear el gráfico de barras
plt.bar(momentos_soleados, frecuencia_relativa, color='skyblue')

# Agregar etiquetas y título
plt.xlabel('Cantidad de momentos soleados')
plt.ylabel('Frecuencia relativa')
plt.title('Distribución de momentos soleados durante 10 días')

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
numero_plantas_sobrevivientes = [0, 1, 2, 3, 4]
frecuencia_relativa = [0.125, 0.25, 0.25, 0.25, 0.125]  # Frecuencia relativa para cada número de plantas sobrevivientes

# Crear el gráfico de barras
plt.bar(numero_plantas_sobrevivientes, frecuencia_relativa, color='skyblue')

# Agregar etiquetas y título
plt.xlabel('Número de plantas sobrevivientes')
plt.ylabel('Frecuencia relativa')
plt.title('Distribución de plantas sobrevivientes en diferentes ubicaciones')

# Mostrar el gráfico
plt.show()

```
